```python
from pyspark.sql import SparkSession
import pandas as pd

spark = SparkSession.builder \
    .appName("Read Excel in PySpark") \
    .config("spark.jars.packages", "com.crealytics:spark-excel_2.12:3.3.0") \
    .getOrCreate()
```

```python
def fetch_data(path_dataset, path_tasks):
    df = spark.read.format("com.crealytics.spark.excel") \
        .option("header", "true") \
            .option("inferSchema", "true") \
                .load(path_dataset)    
    return df
```
